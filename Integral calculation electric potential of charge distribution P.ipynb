

# ** محاسبه انتگرال گیری با روش ذوزنقه‌ای **



```python
import numpy as np
from scipy import integrate

def integrand(x, y):
    return 1 / np.sqrt((40*x - 2*41)**2 + (40*y - 2*41)**2)

result, error = integrate.dblquad(integrand, -1, 1, lambda x: -1, lambda x: 1)

print("Result:", result)
print("Error:", error)
```

# ** محاسبه انتگرال گیری با روش نقطه‌میانی سیمپسون **


```python
import numpy as np
import matplotlib.pyplot as plt

# تابع انتگرالی
def integrand(x, y):
    return 1 / np.sqrt((40*x - 2*41)**2 + (40*y - 2*41)**2)

# تعریف محدوده
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = integrand(X, Y)

# محاسبه انتگرال دوگانه با استفاده از روش سیمپسون
def double_integral_simpson(f, a, b, c, d, nx, ny):
    hx = (b - a) / nx
    hy = (d - c) / ny
    sum = 0
    for i in range(nx):
        for j in range(ny):
            xi = a + i * hx
            xi1 = a + (i + 1) * hx
            yj = c + j * hy
            yj1 = c + (j + 1) * hy
            sum += (hx * hy / 9) * (
                f(xi, yj) + 4 * f((xi + xi1) / 2, yj) + f(xi1, yj) +
                4 * f(xi, (yj + yj1) / 2) + 16 * f((xi + xi1) / 2, (yj + yj1) / 2) +
                4 * f(xi1, (yj + yj1) / 2) + f(xi, yj1) + 4 * f((xi + xi1) / 2, yj1) +
                f(xi1, yj1))
    return sum

result = double_integral_simpson(integrand, -1, 1, -1, 1, 100, 100)
print("Result:", result)

# رسم نمودار
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(label='Integrand Value')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of the Integrand Function')
plt.grid(True)
plt.show()

# رسم نمودار سه‌بعدی
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# تنظیمات نمودار
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Integrand Value')
ax.set_title('Surface Plot of the Integrand Function')
```
